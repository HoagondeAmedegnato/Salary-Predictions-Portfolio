### Let's calculate the Pearson Correlation Coefficient and P-value of some variables

```python
# yearsExperience and salary
pearson_coef, p_value = stats.pearsonr(data2['yearsExperience'], data3['salary'])
print("The Pearson Correlation Coefficient is", pearson_coef, "with a P-value of P =", p_value)
```

```python
# milesFromMetropolis and salary
pearson_coef, p_value = stats.pearsonr(data2['milesFromMetropolis'], data3['salary'])
print("The Pearson Correlation Coefficient is", pearson_coef, "with a P-value of P =", p_value)
```

## Model Development and Evaluation

###  Simple Linear Regression with 'yearsExperience' as a variable

```python
from sklearn.linear_model import LinearRegression
```

```python
# Create the linear regression object
lm = LinearRegression()
lm
```

```python
# Let's see how "yearsExperience" can help us predict salary, by creating a linear function with "yearsExperience" as the predictor variable and the "salary" as the target variable
X = data2[['yearsExperience']]
Y = data3['salary']
```

```python
# Fit the linear model using yearsExperience
lm.fit(X,Y)
```

```python
# Calculating the R^2 (R- squared) as follows:
lm.score(X,Y)
```

```python
# Output a prediction
Y = lm.predict(X)
Y[0]
```

```python
# Know the value of the intercept (b)
lm.intercept_
```

```python
# Know the value of the Slope (a)
lm.coef_
```

```python
## Final estimated linear model:
## X = Predictor Variable(yearsExperience)
## Y = Target Variable(salary)
## Y = aX + b

## Y = 2.01316292X + 91.91919121041059   ## For a Prediction
## Y = 2X + 91
## salary = 2*yearsExperience + 91

## Fitting a Simple Linear Model Estimator
## For X = 0        Y = 91
## For X = 5        Y = 101
## For X = 10       Y = 111
## For X = 15       Y = 121
```

## Conclusion

For yearsExperience and salary, the Pearson Correlation Coefficient is 0.37501269959391104 with a P-value of P = 0.0; this means that we have a strong certainty in the result with a large positive relationship.

For milesFromMetropolis and salary, the Pearson Correlation Coefficient is -0.29766635275577513 with a P-value of P = 0.0; this means that we have a strong certainty in the result, but with a large negative relationship, since this coefficient is close to -1.

The R^2 (R- squared) which is the coefficient of determination, is equal to 1. This is the percentage of variation of the salary variable that is explained by the linear model. This measures how close the data is to the fitted regression line. This value of R^2 indicates that the model performs best.

As the result, we can predict the salary based on years of experience (see simple linear regression method) for the future payroll calculation.
